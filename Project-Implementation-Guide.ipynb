1. Create the GitHub repository for the project.



2. Clone the repository to your development environment.
**git clone https://github.com/sMathujan/AI-Newsletter-Generator.git**

3. Folder structure after clone the repository.



4. Create the conda environment for the project.
**conda create -n news-gen python=3.11**

5. Activate the conda environment.
**conda activate news-gen**

6. Install crewai.
**pip install crewai**

7. Create the crew using CrewAI CLI.
**crewai create newsletter-crew**

8. Folder structure after the crew created by the CrewAI CLI.





9. Re-arrange the folder strcture. Folder structure neede for the project.





10. Fill the tasks.yaml file.

11. Create the agents. Fill the agents,yaml file.

12. Create the tools. 
**tools -> research.py**

13. Put everything together in crew.py file.

14. Pass the inputs to crew in main.py file.

15. Create the newsletter_template.html file.

16. Commands to run the project.

**poetry lock**

**poetry install**

**poetry run newsletter_crew**

17. If you face any package and dependcies problems, try to install them accordingly.
For example, try **poetry add langchain-anthropic**, if not supported, try **pip install langchain-anthropic**

## Web UI for the AI Agent Application

1. Activate the conda environment for the project. 
**conda activate news-gen**

2. Install the streamlit.
**poetry add streamlit**

3. Create the folder structure for the frone-end development.



4. Write the GUI code inside `src\gui\app.py`.

5. Write the code for `set_callback()` function inside the crew.py file.

Sometimes, you may face issues on loading the api_keys from the .env file. In that instances, mention the api_key as the function argument.

6. Command to run the streamlit application.
**streamlit run src\gui\app.py**
